```python
import pandas as pd
df = pd.read_csv('NY-Housing-Data.csv')

df.head()
```

```python
drop_df = df.drop(['BROKERTITLE', 'ADDRESS', 'STATE', 'MAIN_ADDRESS', 'STREET_NAME', 'LONG_NAME', 'FORMATTED_ADDRESS','LATITUDE', 'LONGITUDE'], axis=1)
drop_df
```

```python
df_filtered = drop_df[drop_df['ADMINISTRATIVE_AREA_LEVEL_2'] == 'New York']
df_filtered
```

```python
df_filtered2 = df_filtered[df_filtered['TYPE'].isin(['Condo for sale', 'Co-op for sale', 'House for sale', 'Multi-family home for sale', 'Townhouse for sale'])]
df_filtered2
```

```python
df_dropped2 = df_filtered2.drop(['ADMINISTRATIVE_AREA_LEVEL_2', 'SUBLOCALITY'], axis=1)
df_dropped2
```

```python
df_filtered3 = df_dropped2[df['LOCALITY'].isin(['New York County', 'Kings County', 'Queens County', 'Richmond County', 'Bronx County'])]
df_filtered3
```

```python
df['PROPERTYSQFT'] = df['PROPERTYSQFT'].astype(int)
df['BATH'] = df['BATH'].astype(int)
df_filtered3
```

```python
df_filtered4 = df_filtered3[df_filtered3['PROPERTYSQFT'] != 2184.207862]
df_filtered4
```

```python
tolerance = 1e-6

df_filtered5 = df_filtered4[abs(df_filtered4['BATH'] - 2.373861) > tolerance]

df_filtered5
```

```python
df_final = df_filtered5.reset_index(drop=True)
df_final
```

```python
new_file_name = 'clean_NY_Housing_data.csv'

# Construct the new file path in the current working directory
new_file_path = new_file_name

# Export the DataFrame to the new CSV file
df.to_csv(new_file_path, index=False)
```
